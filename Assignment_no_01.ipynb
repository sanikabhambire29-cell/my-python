```python
#6.	Write a Python program to calculate the factorial of a number, count its digits, and display the reverse of the number.
```

```python
num = int(input("Enter a number: "))

temp = num

fact =  1
i = 1

while i <= num:
    fact = fact * i
    i = i + 1
print("Factorial:", fact)

count = 0
temp2 = temp

while temp2 > 0:
    count = count + 1
    temp2 = temp2 // 10

print("Number of digits:", count)

reverse = 0
temp3 = temp
while temp3 > 0:
    digit = temp3 % 10
    reverse = reverse * 10 + digit
    temp3 = temp3 // 10
print("Reverse number:", reverse)
```

```python
#7.	Write a Python program to generate the Fibonacci series, calculate its sum, and display the largest term generated.
```

```python
n= int(input("Enter number of terms: "))
```

```python
a,b = 0,1
fib_series = []
sum_series = 0

for i in range(n):
  fib_series.append(a)
  sum_series += a
  a,b = b, a + b

print("fibonacci series:",fib_series)
print("sum of the series:",sum_series)

if fib_series:
  print("largest term generated:",max (fib_series))
else:
  print("no terms generated")
```

```python
#11.	Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
text = "Hello World"

reversed_text = text[::-1]

vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for char in text:
    if char.isalpha():  # Check if character is a letter
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Original String:", text)
print("Reversed String:", reversed_text)
print("Number of Vowels:", vowel_count)
print("Number of Consonants:", consonant_count)
```

```python
#24.	Write a Python program to create a matrix using nested lists, display the matrix, and calculate row-wise sums.
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Matrix:")
for row in matrix:
    print(row)

print("\nRow-wise Sums:")
for i in range(len(matrix)):
    row_sum = sum(matrix[i])
    print("Sum of Row", i + 1, ":", row_sum)
```

```python
#25.	Write a Python program to accept N numbers from the user, store them in a list, and display the maximum, minimum, and average values.
```

```python
n = int(input("Enter how many numbers you want to enter: "))

numbers = []

for i in range(n):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

maximum = max(numbers)
minimum = min(numbers)
average = sum(numbers) / n

print("\nList of numbers:", numbers)
print("Maximum value:", maximum)
print("Minimum value:", minimum)
print("Average value:", average)
```
