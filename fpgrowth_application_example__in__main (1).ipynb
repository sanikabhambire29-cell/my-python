```python
import pandas as pd
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# Sample transaction data
```

```python
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Eggs'],
    ['Milk', 'Butter'],
    ['Bread', 'Eggs'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread'],
    ['Butter', 'Eggs']
]
```

```python
# Convert data into one-hot encoding
```

```python
te = TransactionEncoder()
```

```python
te_data = te.fit(transactions).transform(transactions)
```

```python
df = pd.DataFrame(te_data, columns=te.columns_)
```

```python
# Train FP-Growth model
```

```python
frequent_patterns = fpgrowth(df, min_support=0.3, use_colnames=True)
```

```python
print("Frequent Itemsets:")
print(frequent_patterns)
```

```python
# Generate association rules
```

```python
rules = association_rules(
    frequent_patterns,
    metric="confidence",
    min_threshold=0.7
)
```

```python
print("\nFraud Detection Rules:")
```

```python
print(rules[['antecedents', 'consequents', 'support','confidence', 'lift']])
```

```python
import pandas as pd
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
# Supermarket transaction dataset
```

```python
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Eggs'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread', 'Eggs'],
    ['Milk', 'Butter'],
    ['Bread', 'Eggs'],
    ['Milk', 'Bread', 'Butter', 'Eggs'],
    ['Milk', 'Bread'],
    ['Butter', 'Eggs']
]
```

```python
# Convert transactions into 0-1 format
```

```python
te = TransactionEncoder()
```

```python
encoded_data = te.fit(transactions).transform(transactions)
```

```python
df = pd.DataFrame(
    encoded_data,
    columns=te.columns_
)
```

```python
print("Input Dataset:")
print(df)
```

```python
# Train FP-Growth model
```

```python
frequent_itemsets = fpgrowth(
    df,
    min_support=0.3,
    use_colnames=True
)
```

```python
print("\nFrequent Itemsets:")
print(frequent_itemsets)
```

```python
# Generate Association Rules
```

```python
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.6
)
```

```python
print("\nRecommendation Rules:")
```

```python
print(
    rules[
        [
            'antecedents',
            'consequents',
            'support',
            'confidence',
            'lift'
        ]
    ]
)
```
