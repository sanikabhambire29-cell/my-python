```python
class Student:

    school = "ABC College"

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks


    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)
        print("School:", Student.school)

    @classmethod
    def change_school(cls, new_school):
        cls.school = new_school

    @staticmethod
    def grade(marks):
        if marks >= 75:
            return "A"
        elif marks >= 50:
            return "B"
        else:
            return "C"
```

```python
s1 = Student("sanika", 85)
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        Person.__init__(self, name)
        self.roll_no = roll_no

class Result(Student):
    def __init__(self, name, roll_no, marks):
        Student.__init__(self, name, roll_no)
        self.marks = marks
```

```python
r1 = Result("Sanika", 101, 85)
```

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, roll_no):
        self.roll_no = roll_no
        super().__init__(name)

class Result(Student):
    def __init__(self, name, roll_no, marks):
        self.marks = marks
        super().__init__(name, roll_no)

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Marks:", self.marks)

r1 = Result("Sanika", 101, 85)
r1.display()

print("\nInheritance Hierarchy:")
print("Person -> Student -> Result")
```

```python
#14.	Write a Python program to create an abstract class, implement abstract methods, and instantiate child classes.
```

```python
from abc import ABC, abstractmethod
```

```python
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass
```

```python
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width
```

```python
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius
```

```python
rect = Rectangle(10, 5)
```

```python
circle = Circle(7)
```

```python
print("Area of Rectangle:", rect.area())
```

```python
print("Area of Circle:", circle.area())
```

```python
#16.	Write a Python program to overload arithmetic operators, compare objects, and display the results.
```

```python

class Number:
    def __init__(self, value):
        self.value = value


    def __add__(self, other):
        return Number(self.value + other.value)


    def __sub__(self, other):
        return Number(self.value - other.value)


    def __eq__(self, other):
        return self.value == other.value


    def __gt__(self, other):
        return self.value > other.value


    def __str__(self):
        return str(self.value)
```

```python
n1 = Number(20)
```

```python
n2 = Number(10)
```

```python
sum_result = n1 + n2
```

```python
sub_result = n1 - n2
```

```python
print("First Number:", n1)
```

```python
print("Second Number:", n2)
```

```python
print("Addition:", sum_result)
```

```python
print("Subtraction:", sub_result)
```

```python
print("n1 == n2 :", n1 == n2)
```

```python
print("n1 > n2  :", n1 > n2)
```

```python
class Address:
    def __init__(self, city, state):
        self.city = city
        self.state = state

    def display_address(self):
        print("City :", self.city)
        print("State:", self.state)
```

```python
class Employee:
    def __init__(self, emp_id, name, address):
        self.emp_id = emp_id
        self.name = name
        self.address = address  # Composition

    def display_employee(self):
        print("Employee ID:", self.emp_id)
        print("Employee Name:", self.name)
        print("Address Details:")
        self.address.display_address()
```

```python
#23.	Write a Python program to raise a custom exception, handle the exception, and display a user-friendly message.
```

```python
class InvalidAgeError(Exception):
    pass
try:
    age = int(input("Enter your age: "))

    if age < 18:
        raise InvalidAgeError("Age must be 18 or above.")

    print("You are eligible.")

except InvalidAgeError as e:
    print("User-Friendly Message:", e)

except ValueError:
    print("User-Friendly Message: Please enter a valid number.")

finally:
    print("Program execution completed.")
```

```python
#24.	Write a Python program to handle a file-not-found exception, create the missing file, and write default content into it.
```

```python
filename = "sample.txt"
```

```python
try:

    with open(filename, "r") as file:
        content = file.read()
        print("File Content:")
        print(content)

except FileNotFoundError:
    print(f"'{filename}' not found.")
    print("Creating the file with default content...")


    with open(filename, "w") as file:
        file.write("This is the default content of the file.\n")
        file.write("File created automatically because it was missing.")

    print(f"'{filename}' has been created successfully.")

finally:
    print("Program execution completed.")
```

```python
#25.	Write a Python program to create a custom exception class, validate user input, and handle invalid data.
```

```python
class InvalidMarksError(Exception):
    """Raised when marks are outside the valid range"""
    pass

try:
    marks = int(input("Enter marks (0-100): "))

    # Validate input
    if marks < 0 or marks > 100:
        raise InvalidMarksError("Marks must be between 0 and 100.")

    print("Valid marks entered:", marks)

except InvalidMarksError as e:
    print("Custom Exception:", e)

except ValueError:
    print("Error: Please enter a valid integer.")

finally:
    print("Program execution completed.")
```
