```python
#multiple except blocks
#different errors can be handled in different ways using multiple except blocks
```

```python
try:
  num=int("abc")
  result=10/num
except ValueError:
  print("conversion failed")
except ZeroDivisionError:
  print("Division by zero error")
```

```python
try:
  x = 10/0
  result=10/x
except ValueError:
  print("conversion failed")
except ZeroDivisionError:
  print("Division by zero error")
```

```python
#multiple exceptions in one except
#you can handle more than one error type in a single block
```

```python
try:
  x = int("abc")
except (ValueError,TypeError):
  print("value or type error occured")
```

```python
#generic exception
#catch any kind of error when you don't specify the exact type
```

```python
try:
  x = int("abc")
except Exception as e:
  print("Error occured:",e)
```

```python
#else block
#runs only when no error occurs in the try block
```

```python
try:
  x = 10/2
except ZeroDivisionError:
  print("error")
else:
  print("success:",x)
```

```python
#finally block
#runs every time whether an error happens or not often used for cleanup.
```

```python
try:
  x = 10/0
except ZeroDivisionError:
  print("Error occured")
finally:
  print("this always runs")
```

```python
try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error")
else:
    print("Result:", x)
finally:
    print("Execution complete")
```

```python
#raise keyword
# Used to manually create and throw an error.
```

```python
def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18+")
    return "Allowed"

try:
    check_age(15)
except ValueError as e:
    print(e)
```

```python
#Custom Exception
# You can create your own error type for specific situations.
```

```python
class CustomError(Exception):
    pass

def test():
    raise CustomError("This is a custom error")

try:
    test()
except CustomError as e:
    print(e)
```

```python
#assert statement
# Checks a condition; if false, it raises an error.
```

```python
x = -5
try:
    assert x > 0, "x must be positive"
except AssertionError as e:
    print(e)
```

```python
#Nested try-except
# A try-except block placed inside another.
```

```python
try:
    try:
        x = int("abc")
    except ValueError:
        print("Inner error")
        raise
except Exception:
    print("Outer handler caught it")
```

```python
# File handling with exceptions
# Used to handle errors while working with files.
```

```python
try:
    f = open("file.txt", "r")
    data = f.read()
except FileNotFoundError:
    print("File not found")
finally:
    try:
        f.close()
    except:
        pass
```
