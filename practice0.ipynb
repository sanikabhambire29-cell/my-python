```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data = {
    'Week': ['Week1', 'Week2', 'Week3', 'Week4'],
    'Product_A': [120, 135, 150, 145],
    'Product_B': [100, 110, 125, 130],
    'Product_C': [90, 95, 105, 115],
    'Product_D': [80, 85, 90, 100]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.set_index('Week').plot(
    kind='barh',
    figsize=(8, 5)
)
```

```python
plt.title('Product Sales Across Weeks')
plt.xlabel('Sales Units')
plt.ylabel('Weeks')
plt.legend(title='Products')
plt.show()
```

```python
sales_values = df[['Product_A', 'Product_B', 'Product_C', 'Product_D']].values.flatten()
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(sales_values, fill=True, color='skyblue')
plt.title('KDE Plot of Sales Distribution')
plt.xlabel('Sales Units')
plt.ylabel('Density')
plt.show()
```

```python
product_a_sales = df['Product_A']

plt.figure(figsize=(6, 6))
plt.pie(
    product_a_sales,
    labels=df['Week'],
    autopct='%1.1f%%',
    startangle=90
)
```

```python
plt.title('Product_A Sales Contribution Across Weeks')
plt.show()
```
