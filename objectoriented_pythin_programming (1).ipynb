```
# Encapsulation
```

```
# Encapsulation is the process of hiding data using private variables and providing controlled access through methods
```

```
# Encapsulation protects data by restricting direct access
```

```
# Use double underscores (__) to make an attribute private
```

```
class BankAccount:

    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance
```

```
BankAccount(1000).get_balance()
```

```
# Single Inheritance
```

```
# Single inheritance allows one child class to inherit from one parent
```

```
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    pass
```

```
Dog().eat()
```

```
class Student:
    def __init__(self, marks):
        self.marks = marks

class Percentage(Student):
    def show(self):
        print("Percentage:", sum(self.marks) / 5)

Percentage([80, 90, 85, 95, 90]).show()
```

```
# Super
```

```
# super() is used to access parent class methods and constructors
```

```
class Parent:
    def __init__(self):
        print("Parent")

class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child")
```

```
Child()
```

```
# Method overriding
```

```
# Method overriding replaces a parent method in the child class
```

```
class Vehicle:
    def start(self):
        print("Vehicle is starting...")


class Car(Vehicle):
    def start(self):
        print("Car is starting...")
```

```
car = Car()
```

```
car.start()
```

```
# Abstract classes
```

```
# ABC is a class used to force child classes to implement certain methods
# @abstractmethod marks methods that child classes must provide
```

```
# ABC = a rule book
# @abstractmethod = compulsory method
```

```
# An abstract class is a class that cannot be instantiated and is used as a blueprint for other classes
```

```
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):

    def area(self):
        return 78.5
```

```
Circle().area()
```
